```python
# setting up pull all cases
# ran 2000 - 2012

i = 20200001

unique_url = []

while i < 20231300:
    j = str(i)
    k = int(j[4:8])
    if k < 1300:
        unique_url.append(i)
    i = i+1

print(len(unique_url))
```

```python
# pulling all the urls
base_url = f'https://archden.org/tribunal-case-status-search/?caseNumber='

full_url_list = []

for i in unique_url:
    pull_url = base_url + str(i)
    #print(pull_url)
    full_url_list.append(pull_url)
```

```python
print(len(full_url_list))
```

```python
#print(full_url_list[30000])
```

```python
# Iterate Through The 2022 Cases, Number 20,000 - 30,000

case_list = []
case_url_list = []
period_list = []

counter = 0

for i in full_url_list:
    #time.sleep(2)
    html = get(i)
    string_soup = str(Soup(html))
    time.sleep(2)
    location = string_soup.find('Protocol Status')
    #print(location)
    #print('THE LOCATION IS: ', location)
    if location == -1:
        print('No Case')
        print(i)
    else:
        end = int(location) + 75
        status = string_soup[location:end]
        period = string_soup.find('Formal Process')
        time_end = int(period) + 100
        time_length = string_soup[period:time_end]
        #print(status)
        counter = counter + 1
        print(counter)
        period_list.append(time_length)
        case_url_list.append(i)
        #print(counter)
        case_list.append(status)
```
